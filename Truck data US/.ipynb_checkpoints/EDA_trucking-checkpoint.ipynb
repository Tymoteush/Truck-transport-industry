```python
def plot_barsubplots(df,x_col,col_list,x,y,titles_list):
    fig,ax = plt.subplots(x,y,figsize=(14,14))
    ax = ax.flatten()
    plt.suptitle('Hazardous materials transported in the US in 2017')
    for i in range(len(col_list)):
        ax[i].bar(df[x_col],df[col_list[i]],color=['sandybrown','lightcoral','salmon','moccasin','darkkhaki','olivedrab','yellowgreen','darkseagreen','lightgreen'])
        ax[i].title.set_text(titles_list[i])
        plt.subplots_adjust(hspace=0.5)
        plt.setp(ax[i].get_xticklabels(), rotation=30, ha='right')
        
        
plot_barsubplots(hazardous,'Category',['Value','Tons','Ton-miles','Average miles per shipment'],2,2,['Value in billions USD','Tons transported (millions)','Ton-miles (millions)','Average miles per shipment'])
```

### Correlation analysis

Thanks to unified year values among datasets we can perform cross correlation testing between any values from any choosen dataset except dataset containing information about transportation of hazardous materials.

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.scatter(truck_ton_miles['Ton_miles'],total_ton_miles['Ton_miles'])
plt.xlabel('Truck ton-miles')
plt.ylabel('Total ton-miles')
plt.show()
```

We can see that the values are positively correlated. Lets check the pearsons correlation to evaluate the strength of the correlation.

Above result shows us that the correlation is very high and positive(value is 0.88 which is very close to 1) and low p-value indicate significance of the result.

```python
def check_pearsonr_correlation(x,y,description):
    coeff, p_value = stats.pearsonr(x,y)
    plt.scatter(x,y)
    plt.title(description)
    plt.show()
    print(description,'- coefficient: ' + str(coeff),'- p-value: '+ str(p_value))
check_pearsonr_correlation(truck_ton_miles['Ton_miles'],total_ton_miles['Ton_miles'],'Trucking ton-miles vs Total ton miles over the years')
# above function tests 2 variables for correlation and displays correlation coefficient and p-value.
# Pearsons correlation performs also a test of the null hypothesis that the distributions underlying the samples are uncorrelated and normally distributed.
# The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.
```

```python
check_pearsonr_correlation(list_of_tables[0]['Avg_freight_rev_tonemile'],list_of_tables[5]['Operating_revenues'], 'Average freight revenue per tonmile vs total operating revenues')
```

```python
check_pearsonr_correlation(list_of_tables[2]['Diesel_price_per_gallon'],truck_tractors['Number_registered'],'Diesel price per gallon vs registered truck tractors')
```
